4. 자모음의 경우 제거해 주세요 (ㅋㅋㅋㅋㅋ-> '', ㅠㅠ->'')

```python
import pandas as pd
import re
```

```python
eta_df = pd.read_csv(r"질문2.csv", index_col = None ,encoding='ANSI')
```

```python
content = eta_df['내용'].tolist()
for c in content:
    print(c)
    print()
```

```python
norm_content = []
```

```python

for text in content:
    punct = re.compile(r'[\~\'\"\.\,\?\!\:\;\)\(\+]+') 
    norm_text = re.sub(punct, "", text)
    norm_text = re.sub(r'(안녕)+(하세요)?', '', norm_text)
    norm_text = re.sub(r'(조교)[님]?', "", norm_text)
    norm_text = re.sub(r'[ㄱ-ㅎㅏ-ㅣ]{2,}', '', norm_text)
    norm_content.append(norm_text)
```

```python
print(norm_content)
```
