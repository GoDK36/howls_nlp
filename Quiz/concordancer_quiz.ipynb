```python
import os,re


def simple_conc(example_file_path, search_word, result_file_path, result_file_name):
    txt = "E:\Programming\python\창회선배스터디\창회선배 자료\텍스트자료\토지2.txt"

    text = []
    with open(txt, "r", encoding='UTF-8') as filetoread:
        fileread = filetoread.readlines()
        for txt_by_line in fileread:
            text.append(txt_by_line)

    divided_txt = []
    for x in text:
        k = x.split('.')
        divided_txt.append(k)
    divided_txt = sum(divided_txt, [])

    res = []
    for x in divided_txt:
        if search_word in x:
            res.append(x)

    res_path = result_file_path + result_file_name + ".txt"
    result = open(res_path, 'w', encoding='cp949')
    k = 0
    for i in res:
        result.write(str(k+1) + ' => ' + str(i).strip() + '\n')
        k += 1
    result.close()
    os.startfile(res_path)

    print(res)
search_word = input("찾고 싶은 문장에 포함될 단어를 입력해주세요.")

simple_conc(example_file_path, search_word, result_file_path, result_file_name)
```

```python
# 문제1 토지 파일을 온점(".") 기준으로 나누어 읽어서 리스트형 변수 orig_txt에 저장하시오

# 경로 저장
txt = "E:\Programming\python\창회선배스터디\창회선배 자료\텍스트자료\토지2.txt"

# 파일 읽기
with open(txt, "r", encoding='UTF-8') as filetoread:
    fileread = filetoread.read()
    # 온점 기준으로 나누기
    orig_txt = fileread.split('.')

print(orig_txt[:10])
```

```python
# 문제2 토지 파일에서 단어를 입력받아 입력받은 단어가 포함된 문장을 출력해주는 함수를 작성하시오

def conc(word):
    
    # 토지2 파일 경로 저장
    txt = "E:\Programming\python\창회선배스터디\창회선배 자료\텍스트자료\토지2.txt"

    # 파일 읽기
    with open(txt, "r", encoding='UTF-8') as filetoread:
        fileread = filetoread.read()
        # 온점 기준으로 나누기
        orig_txt = fileread.split('.')

    # 원하는 단어의 문장 찾기
    res = []
    for x in orig_txt:
        if word in x:
            res.append(x)

    print(res)

conc(input('단어'))
```

```python
# 문제3 문제2에서 만든 함수의 출력값을 파일로 만들어 저장하시오

# 결과 파일이 저장될 경로
result = open(r"E:\Programming\python\NLP\HOW'Ls\Quiz\conc_Result.txt", 'w', encoding='cp949')
# 문장 수 카운트
k = 0
# 파일쓰기
for i in conc(input("단어")):
    result.write(str(k+1) + ' => ' + str(i).strip() + '\n')
    k += 1
result.close()

os.startfile(r"E:\Programming\python\NLP\HOW'Ls\Quiz\conc_Result.txt")
```
