```python
content = eta_df['내용'].tolist()
ids = eta_df['id'].tolist()
```

```python
DM = [0]*len(ids) #double major
SM = [0]*len(ids) #sub major
UM = [0]*len(ids) #unified major??
MD = [0]*len(ids) #major deepening??
UK = [0]*len(ids) #unknown
```

```python
len(UM)
```

```python
UM_WORD = ["융합", "융합전공"
"BRICS", "브릭스",
"동북아외교통상", "동북아 외교 통상",
"외국어로서의한국어교육","외국어로서의 한국어 교육",
"언어와공학","언어와 공학",
"디지털인문한국학",
"유럽연합", "EU",
"문콘", "문화콘텐츠학","문화 콘텐츠",
"국가리더","국가 리더",
"융소","융복합소프트웨어", "융복합 소프트웨어",
"세문경", "세계문화예술경영", "세계 문화 예술 경영",
"정보기록학",
"광역특화"]
```

```python
#융합 전공을 추출하는 방식을 바꿔보았습니다.

cnt = 0
for i, doc in enumerate(content):
    for j in UM_WORD:
        if j in doc:
            try:
                UM[i] = 1
            except:
                print(i)
            print(doc)
            cnt += 1
            break

print('여러 전공 이름 추가 방식', cnt)

cnt = 0
for i, doc in enumerate(content):

    if '융합' in doc or '융합전공' in doc:
        UM[i] = 1
        cnt += 1

print('기존방식', cnt)
```

```python
for i, doc in enumerate(content):
    if '이중' in doc:
        DM[i] = 1
    if '부전' in doc:
        SM[i] = 1
    if '전공심화' in doc or '전심' in doc:
        MD[i] = 1

    for j in UM_WORD:
        if j in doc:
            UM[i] = 1
            cnt += 1
            break
    if DM[i] == 0 and SM[i]==0 and MD[i]==0 and UM[i]==0:
        UK[i]==1
```

```python
raw_data = {'이중전공': DM,
            '부전공': SM,
            '전공심화': MD,
           '융합전공': UM,
           'UNK': UK,
           'id': ids}

data = pd.DataFrame(raw_data)
```
